# 处理、操作

```python
# 行为总数分布
count1,count2,count3,count4 = 0,0,0,0

for i in df:
  count1 += i[i[2]==1].count()
  count2 += i[i[2]==2].count()
  count3 += i[i[2]==3].count()
  count4 += i[i[2]==4].count()

print(count1)
print(count2)
print(count3)
print(count4)
```

```python
df1 = pd.DataFrame(data=[count1[2],count2[2],count3[2],count4[2]],
                   index=['浏览','收藏','加购物车','购买'],
                   columns=['行为数量'])
df1
```

```python
df1['行为比例'] = df1['行为数量']/(df1['行为数量'].sum())
df1
```

```python
df1['漏斗'] = 1-abs(df1['行为数量'].pct_change(1))
df1
```
